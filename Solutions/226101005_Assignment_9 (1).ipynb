```python
import numpy as np
import pandas as pd
import math

# Read training and test files
df_train = pd.read_csv('Train_file.csv', header=None)
df_test = pd.read_csv('Test_file.csv', header=None)

# Split data into features and labels
X_train = df_train.iloc[:, :-1]
y_train = df_train.iloc[:, -1]
X_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]

# Extract features for class 1 and class 0
x_1 = X_train[y_train == 1]
x_0 = X_train[y_train == 0]

# Compute mean
mu_1 = x_1.mean()
mu_0 = x_0.mean()

# Compute covariance matrix
C1 = x_1.cov()
C0 = x_0.cov()

# Compute discriminant function parameters
P_x1 = len(x_1) / len(X_train)
P_x0 = len(x_0) / len(X_train)

k1 = math.log(P_x1 / P_x0)
mu1_T = mu_1.transpose()
mu0_T = mu_0.transpose()
c1_inv = np.linalg.inv(C1)
c0_inv = np.linalg.inv(C0)

k2 = np.dot(np.dot(mu1_T, c1_inv), mu_1)
k3 = np.dot(np.dot(mu0_T, c0_inv), mu_0)
k4 = -0.5 * (math.log(np.linalg.det(C1) / np.linalg.det(C0)))
k5 = np.dot(c1_inv, mu_1) - np.dot(c0_inv, mu_0)

# Initialize dictionary to store discriminant function values
dic = {}

# Compute discriminant function value for each test example
for index, row in X_test.iterrows():
    row = row.values.reshape(60, 1)
    row1 = row.transpose()
    g_x = k1 - 0.5 * (k2 - k3) + k4 + np.dot(row1, k5) - 0.5 * (np.dot(np.dot(row1, (c1_inv - c0_inv)), row))
    g_x = g_x.item()
    if g_x >= 0:
        dic[g_x] = 1
    else:
        dic[g_x] = 0

# Dictionary contains discriminant function value for each training example and its final value 1 or 0
print(dic)
```
