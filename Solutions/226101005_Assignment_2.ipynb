```python
import numpy as np #4a
import matplotlib.pyplot as plt

# Set the center and axis lengths of the ellipse

center = [-10, 20]
semi_major_axis = 150
semi_minor_axis = 100

# Generate 1000 points within the ellipse

n = 1000
theta = np.random.rand(n) * 2 * np.pi
radius = semi_major_axis * np.sqrt(np.random.rand(n))
x = center[0] + radius * np.cos(theta)
y = center[1] + radius * semi_minor_axis / semi_major_axis * np.sin(theta)
points = np.column_stack((x, y))

# Plot the points

plt.scatter(points[:, 0], points[:, 1])

# Plot the ellipse using the plot function

theta_grid = np.linspace(0, 2*np.pi, 100)
x_ellipse = center[0] + semi_major_axis*np.cos(theta_grid)
y_ellipse = center[1] + semi_minor_axis*np.sin(theta_grid)
plt.plot(x_ellipse, y_ellipse,'r', linewidth=2)

# Show the plot

plt.show()
```

```python
import numpy as np

# N = 5000, m = 400
N = 5000
m = 400

# Randomly generate frequency values
frequencies = np.random.randint(1, high=100, size=m)

# Normalize the frequencies to create a probability distribution
probabilities = frequencies / sum(frequencies)

# Create a weighted dataset
dataset = list(zip(range(1, m+1), probabilities))

# Evaluate the weighted AM
am = sum([x[0] * x[1] for x in dataset])
print("Weighted AM: ", am)

# Evaluate the weighted GM
gm = (np.prod([x[0] ** x[1] for x in dataset])) ** (1/m)
print("Weighted GM: ", gm)

# Evaluate the weighted HM
hm = sum([x[1] * np.log(x[0]) for x in dataset])
print("Weighted HM: ", hm)

# Evaluate the weighted median
dataset.sort(key=lambda x: x[1], reverse=True)
cumulative_prob = 0
for i in dataset:
    cumulative_prob += i[1]
    if cumulative_prob >= 0.5:
        median = i[0]
        break
print("Weighted Median: ", median)

# Evaluate the mode
mode = max(dataset, key=lambda item: item[1])[0]
print("Weighted Mode: ", mode)
```

```python
import numpy as np

def weighted_summary_stats():
    # data should be a list of tuples, where each tuple is of the form (x_i, w_i)
    # and w_i is the weight associated with x_i
    data = []
    n = int(input("Enter the number of data points: "))
    for i in range(n):
        x, w = map(float, input(f"Enter x and w for data point {i+1} separated by a space: ").split())
        data.append((x, w))
    x = [d[0] for d in data]
    w = [d[1] for d in data]
    
    # Weighted arithmetic mean (AM)
    am = np.average(x, weights=w)
    print(f"Weighted arithmetic mean (AM): {am}")
    
    # Weighted geometric mean (GM)
    gm = np.exp(np.average(np.log(x), weights=w))
    print(f"Weighted geometric mean (GM): {gm}")
    
    # Weighted harmonic mean (HM)
    hm = np.sum(w) / np.sum([w[i]/x[i] for i in range(len(x))])
    print(f"Weighted harmonic mean (HM): {hm}")
    
    # Weighted median
    sorted_data = sorted(data, key=lambda d: d[0])
    weight_sum = 0
    for d in sorted_data:
        weight_sum += d[1]
        if weight_sum >= 0.5:
            weighted_median = d[0]
            break
    print(f"Weighted median: {weighted_median}")

weighted_summary_stats()
```
